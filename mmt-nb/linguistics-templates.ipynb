## Crear una plantilla a partir de una definición anotada previamente

Este paso es opcional. Es un ejemplo de cómo tomar una definición ya anotada y convertirla en un template. Se usa la función `package.get_custom_table()` para obtener las anotaciones. Esa función requiere cincto parámetros: el ID de un Google Sheets, el nombre de ls hoja con los datos, celda de inicio de los datos, celda final de los datos (puede ser una celda mayor incluso), y un booleano (False o True) que se puede cambiar si se desea trasponer los datos. La función devuelve un *dataframe* de pandas.

```python
sheet_datos_anotados = '1YyfFSL2-yTdw5LDcmu-AiXi83SZgzbprQyMV2NYvuyQ'
dimension = 'a-f-pos'
```

```python
nivel_basico = package.get_custom_table(sheet_datos_anotados, dimension, 'A1', 'J10000', False)
```

Revisar el número de filas y columnas del dataframe:

```python
nivel_basico.shape
```

```python
nivel_basico.head()
```

Ver los IDs de las oraciones anotadas. Hay que notar el número de cada una (el número que sigue a **SP-**).

```python
nivel_basico.id.unique()
```

```python
num_anotacion = 5
```

```python
def_anotada = nivel_basico[nivel_basico.id=='Corpus-Lexicon-MammutPrime-SP-{0}-0-0'.format(num_anotacion)]
```

Ver la definición anotada:

```python
def_anotada
```

Para convertir esta anotación en una plantilla hacen falta dos pasos. El primero es seleccionar y convertir en una variable uno o varios entities. Para ello se utiliza la función `span_calculator()`. Esta función requiere de tres parámetros: un *dataframe* con las anotaciones (en este caso, es el dataframe que se muestra arriba), una *lista* (entre corchetes) con los entities que se van a convertir en variables, y otra *lista* con los nuevos nombres de las variables.

```python
package.span_calculator(def_anotada, ['Tiempo', 'periodo'], ['XXXXXXXX', 'YYYYYYYY'])
```

Como se puede observar, la función recalcula todos los spans en el cuerpo de anotaciones con la longitud nueva de las variables. Faltaría cambiar el número del ID de la plantilla, para que no se corresponda con el de otra plantilla. Eso lo haremos con la función `id_replace()`. Esta función requiere de dos parámetros: un *dataframe* de pandas y un número que indica el nuevo ID de la anotación. **Cabe destacar que la sustitución de entities no es perfecta y que una vez obtenido el resultado hay que editar algunas cosas en Google Sheets**. En este ejemplo, en el entity *el periodo de duración de algo* hay que sustituir a mano (en Google Sheets) la palabra *periodo* por YYYYYYYY.

```python
package.id_replace(def_anotada, 99)
```

Arriba se puede apreciar el cambio en el número del ID de la oración anotada.

Los ejemplos de arriba no se han almacenado en ninguna parte. Para guardar los cambios hace falta asignar a una variable el resultado de las funciones. Vamos a guardar el resultado de ambas funciones en la variable `resultados`. Para ello, vamos a usar una función dentro de la otra, y asignaremos el dataframe resultante a la variable:

```python
resultados = package.id_replace(package.span_calculator(def_anotada, ['Tiempo', 'periodo'], ['XXXXXXXX', 'YYYYYYYY']), 99)
```

La plantilla nueva ha sido almacenada en `resultados`, como puede verse a continuación:

```python
resultados
```

Finalmente, agregamos el resultado al Google Sheets que queramos.

```python
package.set_custom_table(sheet_datos_anotados, resultados, 'test')
```

Otras dos alternativas son:  
- **Seleccionar la tabla con el mouse, dejando por fuera los títulos, y usar Ctrl-C para copiar los datos y pegargos con Ctrl-V en un Google Sheet**: va a pegar los datos con el índice -la primera columna de la tabla-. 
- **Guardar los resultados en un archivo CSV en la carpeta de notebooks usando el método `to_csv()`**: el primer parámetro es el nombre del archivo que se guardará.

```python
resultados.to_csv('resultados.csv', sep=',', index=False, header=None)
```

El archivo se almacena en la carpeta de notebooks, y es visible desde el Labs. Se puede descarcar haciendo clic sobre el nombre del mismo con el segundo botón del mouse.

Una vez descargado, el CSV se puede subir desde un Google Sheets con **File -> Import...**

Estos datos de anotación ya están listos para ser incorporados en el excel donde se almacenarán las plantillas. Hay que recordar que la plantilla debe ser consistente en todos los niveles de anotación. Esto quiere decir que **si una plantilla tiene dos variables (XXXX y ZZZZ), esas mismas dos variables deben aparecer en todos los niveles de anotación de la definición**. Otro aspecto importante a tener en cuenta es que las plantillas **no deben requerir de restituciones de elementos implícitos**, **ni tener solapamientos en las entidades o eventos**. Además, una plantilla determinada debe tener **el mismo número de ID en todos los niveles** para asegurar que el framework la reconozca como un todo.

## Crear una definición anotada a partir de una plantilla

El proceso de usar una plantilla para crear a partir de ella una definición es igual al anterior, sólo que se parte de la plantilla y se asignas valores léxicos reales a las variables. Se usa la función `package.get_custom_table()` para obtener datos de un Google Sheets con las plantillas:

```python
sheet_template = '1a_zJvEh6HuRKnOlGQsh-khD-6azMBUhYmBaOlyoyLC0'
dimension = 'a-t-pos'
```

```python
templates = package.get_custom_table(sheet_template, dimension, 'A1', 'J10000', False)
```

```python
templates.shape
```

```python
templates
```

```python
templates.id.unique()
```

Vamos a seleccionar la plantilla # 99:

```python
num_template = 1
```

```python
template = templates[templates.id=='Corpus-Lexicon-MammutPrime-SP-{0}-0-0'.format(num_template)]
```

Ver la plantilla anotada:

```python
template
```

Con la nueva plantilla seleccionada, usamos las funciones `id_replace()` y `span_calculator()` para obtener la nueva definición anotada a partir de la plantilla. En este caso, en la función `span_calculator()`, **se pone primero la lista de variables, y seguidamente la lista de los nuevos valores**. Se selecciona también el número del ID que tendrá la nueva definición anotada. Recordamos que los parámetros de la función `span_calculator()` son, en este caso, *template de un nivel*, *lista de variables*, *lista de valores nuevos*.

```python
nueva_definicion = package.id_replace(package.span_calculator(template, ['XXXX', 'YYYY'], ['extensión temporal', 'tiempo']), 12)
```

A continuación se puede observar la nueva definición construida a partir de la plantilla.

```python
nueva_definicion
```

Nótese que en uno de los entities quedó la variable YYYYYYYY por cambiar. Esto se puede hacer directamente sobre el Google Sheets. No hace falta recalcular el span, ya que el que está en la anotación es correcto.
